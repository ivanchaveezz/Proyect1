```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('universal_top_spotify_songs.csv')
country_danceability = data.groupby('country')['danceability'].mean().reset_index()
country_danceability = country_danceability.sort_values(by='danceability', ascending=False)

plt.figure(figsize=(20, 12))
plt.barh(country_danceability['country'], country_danceability['danceability'], color='skyblue')
plt.title('Danceability by Country (Sorted by Danceability)')
plt.xlabel('Mean Danceability')
plt.ylabel('Country')

plt.show()
```
